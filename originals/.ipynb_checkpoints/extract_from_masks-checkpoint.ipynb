```python
def extract_from_masks(mask_files,data_file):
    #given a data file and a set of mask files, extract timeseries from the data 
    #for each mask. Returns a dictionary with the key corresponding to the mask
    #(derived from the file name) and the value as a numpy array that's TR x voxels
    output = [apply_mask(data_file,m) for m in mask_files] 
    masknames = [op.splitext(op.splitext(op.basename(x))[0])[0] for x in mask_files]
    output = dict(zip(masknames,output))
    return(output)    
```

```python
from glob import glob 
analysis_dir = os.path.expanduser('/Users/jason/Documents/neuro_econ/observe/analysis/observe_man2/')

os.chdir(analysis_dir)
subids = glob('000*') #individual subject directories where copes will be found


# badsubs = ['0001424']

# for bs in badsubs:
#     subids.pop(subids.index(bs))
    
print 'found %d subjects:' %len(subids)
print subids

```

```python
#set up directories and find images

img_name = 'cope15' #image name to extract (minus file extension)

template = '{subject_id}/model/smoothed/run_1'


cope_files=[]
for s in subids:
    
    temp = template.replace('{subject_id}',s)
    statdir = os.path.join(analysis_dir,temp)
    


    os.chdir(statdir)
    copefile= os.path.join(statdir,glob(img_name+'.nii.gz')[0])
    cope_files.append(copefile)

print cope_files
```

```python
#now extract everything 
data_dir = os.path.expanduser('~/Documents/big_data/neuro_econ/observe/data') #base directory to find masks



# masks = ['l_nacc','l_insula','l_caudate','l_posteriorcingulate','l_precuneus',
#          'r_nacc','r_insula','r_caudate','r_posteriorcingulate','r_precuneus',
#          'anterior_vmpfc','central_vmpfc','decision_value_vmpfc','ventral_vmpfc','neurosynth_decision_acc'] #these have to match the filenames (minus the extension)


masks = ["roi--23-11-64", "roi--44-2-46", "roi--47-11-23", "roi--53--49-43",
         "roi-24-45--15", "roi-34-54--13", "roi-48-25-27", "roi-58--53--14"]

#timeseries, or a single contrast image?
timeseries = False 


if not timeseries:
    #dataframe for storing all the info
    alldata = pd.DataFrame(data={'subject': subids})
else:
    #this is for doing the timeseries 
    alldata = pd.DataFrame(index=[],columns=['subject','vol',masks[0]])

for masknum,m in enumerate(masks):
    
    print 'extracting mask %s, (%d of %d)...' %(m,masknum+1,len(masks))
    
    mask_files =[]
    
    for s in subids:
#         print 'subject %s' %s
        subdir = os.path.join(data_dir,s,'masks','network')

        maskfile = findfiles(subdir,m+'.nii.gz')['files'][0]
        mask_files.append(maskfile)
        
    
    result = extract_from_masks(mask_files,cope_files,timeseries)
    
    if not timeseries:    
        result['subject'] = subids
        alldata = pd.merge(alldata,result,on='subject')
    else:
        
        subdata = pd.DataFrame(index=[],columns=['subject','vol',m])
        
        for i,ts in enumerate(result):
            ids= pd.Series(np.repeat(subids[i],len(ts),0))
            temp = pd.DataFrame(data=ids,columns=['subject'])
            temp['vol'] = np.arange(len(ts))+1
            temp[m] = ts

            subdata = pd.concat([subdata,temp],axis=0)
        
        if masknum==0:
            alldata = pd.concat([alldata,subdata],axis=0) #first mask, just add the data
        else:
            alldata = pd.merge(alldata,subdata,on=['subject','vol']) #subsequent masks, make a wide dataset
    




alldata.head(50)
```
